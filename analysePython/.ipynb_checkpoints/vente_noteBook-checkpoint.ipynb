ANALYSE DE VENTES

Préparation des données

Avant d'entamer l’analyse, une phase de préparation des données a été nécessaire. Tout d’abord, la colonne 'Date', initialement au format texte (string), a été convertie en format datetime. Cette conversion facilite la manipulation temporelle des données, notamment pour extraire des informations comme le mois, ce qui a permis la création d'une nouvelle colonne 'mois' utile pour l’analyse des ventes en fonctions des différentes période de l'année.

Ensuite, la colonne des prix de vente, qui contenait le symbole '€', a été nettoyée. Ce caractère empêchait les opérations numériques, il a donc été supprimé et la colonne a été convertie en type float pour permettre les calculs.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data_frame=pd.read_csv('../data/ventes.csv',header=0, sep=";")
df=data_frame.copy()

#préparation du data frame
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df['mois'] = df['Date'].dt.month
df['Amount'] = df['Amount'].str.replace('€', '').str.replace(' ','').str.replace(',','.').astype(float)
#print(df.isnull().sum())
print(df.info())
```

```python
#montant de vente par zone geographique
df_geo=df.groupby('Geography',as_index=False)['Amount'].sum()
plt.figure(figsize=(6, 4))
plt.bar(df_geo['Geography'],df_geo['Amount'] )
plt.title('Vente par region')
plt.xlabel('Région')
plt.ylabel('total vente')
plt.show()
```

![vente par région](../assets/vente_region.png)

Analyse des ventes par région

Ce premier graphique illustre le total des ventes par région. L’objectif était d’identifier quelle région génère le chiffre d’affaires le plus élevé. Il en ressort clairement que la région 'India' se distingue avec le montant de ventes le plus important, surpassant ainsi les autres régions.

```python
#produit les plus vendu par region
df_product=df.groupby(['Product', 'Geography'], as_index=False)['Amount'].sum()
for c in df_product['Geography'].unique():
    plt.figure(figsize=(6, 4))
    sns.barplot(data= df_product[df_product['Geography'] == c],
                x='Product',
                y='Amount',
                )
    plt.xticks(rotation=90) 
    plt.title(f'produit vendu en {c}')
    plt.xlabel('produit')
    plt.ylabel('total vente')
    plt.show()
```

Analyse des produits les plus vendus par région

Les graphiques ci-dessus présentent les produits ayant généré le plus de chiffre d'affaires (C.A.) dans chaque région. Voici les principales observations :
Australia : Le produit le plus vendu est Peanut Butter Cubes, suivi de très près par Caramel Stuffed Bars.
Canada : Une nette domination de Manuka Honey Choco se dégage.
India : Deux produits ressortent : 70% Dark Bites en tête, suivi de Caramel Stuffed Bars.
New Zealand : Le produit dominant est 99% Dark & Pure.
UK : White Choc se distingue comme le produit le plus rentable, suivi de Fruit & Nut Bars.
USA : 70% Dark Bites s’impose à nouveau, devenant ainsi le seul produit à apparaître en tête dans plusieurs régions.

Ces résultats permettent d’identifier les préférences régionales et les produits phares qui contribuent le plus au chiffre d’affaires global.

```python
#vendeur ayant le montant de vente le plus elevé
df_vendeur=df.groupby('Sales Person',as_index=False)['Amount'].sum()
plt.figure(figsize=(6, 4))
plt.bar(df_vendeur['Sales Person'],df_vendeur['Amount'] )
plt.xticks(rotation=90) 
plt.title('total des ventes des vendeur')
plt.xlabel('Vendeur')
plt.ylabel('total vente')
plt.show()
```

![vente par vendeur](../assets/vente_vendeur.png)

Analyse des vendeurs ayant les plus vendus

cette analyse permet de savoir quel est le vendeur ayant le plus grand total de vente à l'année. Le graphe ci dessous nous montre que ce vendeur est le vendeur 'Andria Kimpton'.
Mais nous pouvons ajouter un plus à ce graphe en y ajoutant l'indicateur de la région ainsi nous auront les région dans lequel chaque vendeur a fait ces meilleures vente.

```python
#les vendeurs ayant le plus de vente par region
df_vendeur_region=df.groupby(['Sales Person','Geography'],as_index=False)['Amount'].sum()
plt.figure(figsize=(7, 4))
sns.barplot(data= df_vendeur_region,
            x='Sales Person',
            y='Amount',
            hue='Geography'
            )
plt.xticks(rotation=90) 
plt.title('vente total des vendeur par région')
plt.xlabel('vendeur')
plt.ylabel('total vente')
plt.show()
```

Analyse des vendeurs ayant les plus vendus par région

A l'aide de ce nouveau visuel nous avons maintenant de voir dans quel région un vendeur est le performant. Si on reprend le vendeur qui a fait les meilleures vente à l'année qui est 'Andria Kimpton' on remarque une forte concentration de ces ventes en 'Australia'.

```python
#periode de l'année avec le plus de vente
df_mois=df.groupby('mois',as_index=False)['Amount'].sum()
plt.figure(figsize=(6, 4))
plt.bar(df_mois['mois'],df_mois['Amount'] )
plt.title('periode de l année avec le plus de vente')
plt.xlabel('mois')
plt.ylabel('total vente')
plt.show()
```

![vente par mois](../assets/vente_mois.png)

Analyse des meilleurs vente selon la période de l'année

Le graphe ci-dessus nous montre le mois où il y'a eu les meilleures vente, Les chiffres de 1 à 5 correspondent au numéro du mois donc de Janvier à Mai. On remarque que le meilleur mois est le mois de Janvier suivit par le mois de Mars.
A cette analyse on peut ajouter comme précédement le facteur région pour obtenir les regions qui où il y'a eu les meilleures vente en fonction du mois.

```python
#periode de l'année avec le plus de vente par region
df_mois_region=df.groupby(['mois','Geography'],as_index=False)['Amount'].sum()
plt.figure(figsize=(6, 4))
sns.barplot(data= df_mois_region,
            x='mois',
            y='Amount',
            hue='Geography'
            )
plt.title('periode de l année avec le plus de vente par région')
plt.xlabel('mois')
plt.ylabel('total vente')
plt.show()
```

Analyse des meilleurs vente par région selon la période de l'année

Ce nouveau graphique nous permet de savoir maintenant en quel mois il y'a le plus de demande par région. Si nous reprenons le mois de Janvier qui est le mois qui a les meilleures vente on constate que pendant ce mois la région où il y'a le plus de demande est 'Australia'.

Ce notebook a permis d’explorer et d’analyser en profondeur les ventes réalisées à travers différentes régions et pérodes de l'année. Après une phase de préparation des données (nettoyage, conversion de types, et création de nouvelles variables), plusieurs analyses visuelles ont été menées afin de répondre aux principales questions commerciales.
Les visualisations ont permis notamment de :

Identifier les régions les plus rentables, avec notamment India comme leader en termes de chiffre d’affaires.

Distinguer les produits les plus performants par région, révélant des préférences marquées selon les zones géographiques.

Mettre en évidence des produits phares, comme 70% Dark Bites, qui se démarquent par leur succès dans plusieurs pays.

Ces résultats offrent des pistes concrètes pour orienter les décisions marketing, telles que le ciblage de certains produits selon les régions, ou le renforcement de la distribution dans les zones les plus performantes.
